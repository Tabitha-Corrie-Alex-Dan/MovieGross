```python
import pandas as pd
import json
```

```python
year = ['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016', '2017', '2018', '2019', '2020', '2021']
season = ["Spring","Summer","Fall","Winter"]

years = len(year)
```

```python
imdb = []
```

```python
i=0


for y in year:
    i = i+1
    url = 'https://www.boxofficemojo.com/season/by-year/'+y+'/?grossesOption=calendarGrosses'
    imdb.append(pd.read_html(url))
    print(imdb)
    
```

```python
# imdb_df = pd.DataFrame()

for i in range(1,22):
    imdb[i] = imdb[i][0].assign(Year=year[i])
```

```python
movies = []
for i in range(1,22):
    movies.append(imdb[i])



imdb_df = pd.concat(movies)
    
```

```python
imdb_df["Cumulative Gross"] = imdb_df["Cumulative Gross"].replace('[\$\,\.]',"",regex=True).astype(int)
imdb_df["Average"] = imdb_df["Average"].replace('[\$\,\.]',"",regex=True).astype(int)
imdb_df["Gross"] = imdb_df["Gross"].replace('[\$\,\.]',"",regex=True).astype(int)
```

```python
sum_df = imdb_df.groupby(['Season']).sum()
```

```python
def format(x):
        return "${0:,.2f}".format(x/1000)
    
sum_df['Cumulative Gross'] = sum_df['Cumulative Gross'].apply(format)
sum_df['Gross'] = sum_df['Gross'].apply(format)
sum_df['Average'] = sum_df['Average'].apply(format)
```
